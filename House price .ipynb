```python
columns_numeric = list(df_train.select_dtypes(exclude='object'))
columns_cat = list(df_train.select_dtypes(include='object'))
```

```python
columns_numeric.remove('Id')
columns_numeric.remove('SalePrice')
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer_num = SimpleImputer(strategy='median')
imputer_num.fit(df_train[columns_numeric])
df_train[columns_numeric] = imputer_num.transform(df_train[columns_numeric])

imputer_cat = SimpleImputer(strategy='most_frequent')
imputer_cat.fit(df_train[columns_cat])
df_train[columns_cat] = imputer_cat.transform(df_train[columns_cat])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(df_train[columns_cat])
temp = pd.DataFrame(data = ohe.transform(df_train[columns_cat]).toarray(),columns= ohe.get_feature_names())
```

```python
columns_final = columns_numeric+list(ohe.get_feature_names())
```

```python
df_train = pd.concat([df_train,temp],axis=1)
df_train
```

```python
df_train[columns_final].head()
```

```python
df_test[columns_numeric] = imputer_num.transform(df_test[columns_numeric])
df_test[columns_cat] = imputer_cat.transform(df_test[columns_cat])

temp = pd.DataFrame(data=ohe.transform(df_test[columns_cat]).toarray(), columns=ohe.get_feature_names())
df_test = pd.concat([df_test, temp], axis=1)
```

```python
x = df_train[columns_final]
y = df_train['SalePrice']
```

```python
from xgboost import XGBRegressor
```

```python
model_XGB = XGBRegressor(base_score=0.5, booster='gbtree', callbacks=None,
             colsample_bylevel=0.5, colsample_bynode=1, colsample_bytree=0.4,
             early_stopping_rounds=None, enable_categorical=False,
             eval_metric=None, feature_types=None, gamma=0, gpu_id=-1,
             grow_policy='depthwise', importance_type=None,
             interaction_constraints='', learning_rate=0.02, max_bin=256,
             max_cat_threshold=64, max_cat_to_onehot=4, max_delta_step=0,
             max_depth=5, max_leaves=0, min_child_weight=2,
             monotone_constraints='()', n_estimators=1400, n_jobs=0,
             num_parallel_tree=1, predictor='auto', random_state=0)
model_XGB.fit(x,y)
```

```python
yp_xg = model_XGB.predict(df_test[columns_final])
```

```python
df_submission = pd.DataFrame({'Id':df_test['Id'], 'SalePrice':yp_xg})
df_submission.to_csv('xg.csv', index=False)
```
